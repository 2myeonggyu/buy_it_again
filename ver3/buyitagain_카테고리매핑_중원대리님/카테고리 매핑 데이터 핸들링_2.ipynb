# # 추가, <u>하나의 물리카테고리(배송2.0)는 표준 카테고리(라이브딜) 하나에 mapping되었나?<u/>

```python
# 예를 들어, 다음과 같은 물리카테고리의 mapping 정보를 살펴보자.

cate_deli.dcate_cd.unique()[0]
```

```python
# 이러한 표준 카테고리에 mapping 되었구나! 

cate_deli[cate_deli.dcate_cd == cate_deli.dcate_cd.unique()[0]].id1.values
```

```python
cate_deli.dcate_cd.nunique()
```

```python
# 그렇다면 전체적인 mapping 정보를 살펴보자.
id_1231 = {}

for i in range(cate_deli.dcate_cd.nunique()):
    id_1231[i] = cate_deli[cate_deli.dcate_cd == cate_deli.dcate_cd.unique()[i]].id1.values # category id 매핑정보를 얻는다
```

```python
id_1231
```

```python
# 252index의 물리카테고리의 mapping 정보 살펴보면 
cate_deli.dcate_cd.unique()[252]
```

```python
# 이렇게 많은 표준카테고리로 mapping 되어있는것을 살필 수 있다. - 하지만 이건 예외적인 경우다.
cate_deli[cate_deli.dcate_cd == cate_deli.dcate_cd.unique()[252]].id1.values
```
