## # prev - cur 데이터 merge한다 : 매핑함으로, 통계량이 어느정도 변하는지 살펴볼거야

## # 카테고리 매핑으로 인해 통계량들의 업데이트 된 정보를 보고싶다(rcp의 차이라던지, 고객수의 차이 등)

```python
# 이렇게 하면 되긴 하지만.. 경고 밣생

prev_cur_common['e'] = pd.Series(prev_cur_common.rcp_y - prev_cur_common.rcp_x, index=prev_cur_common.index)
```

```python
# 이렇게 해도.. 되긴 하지만.. 경고 밣생

prev_cur_common["rcp_diff"] = prev_cur_common.rcp_y - prev_cur_common.rcp_x
```

```python
# 이런식으로 하는게 적당하다네!

df1 = prev_cur_common.assign(rcp_diff=pd.Series(prev_cur_common.rcp_y - prev_cur_common.rcp_x).valeus )
```

### # rcp_diff 생성

```python
# values 따로 안해도 되겠다.

df2 = prev_cur_common.assign(rcp_diff=prev_cur_common.rcp_y - prev_cur_common.rcp_x)
```
