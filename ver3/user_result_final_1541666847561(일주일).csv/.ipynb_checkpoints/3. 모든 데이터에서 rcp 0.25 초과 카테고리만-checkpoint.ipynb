```python
### 3개월치의 훈련데이터!  (단위 - 1주일)

chunk = {}
data = pd.DataFrame()
for i in range(200):
    
    chunk[i] = pd.read_csv('part-{:05d}-15df032a-a04c-4e0c-aaaf-46c836ed77f3-c000.csv'.format(i))
    data = data.append( chunk[i][chunk[i].rcp > 0.25] )
```
