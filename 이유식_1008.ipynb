# 함수정의

```python
def func_lamb(train, test, test_id, current, alpha, beta, tmean):
    t_ = (current - train[train.m_id == test_id].iloc[-1].biz_dt).days
    t_purch = (train[train.m_id == test_id].iloc[-1].biz_dt - train[train.m_id == test_id].iloc[0].biz_dt).days
    k = train.biz_dt[train.m_id == test_id].shape[0]
    if k <= 1 :
        tmean_u = tmean
    else:
        tmean_u = train.biz_dt[train.m_id == test_id].diff().mean().days
    lamb = (k+alpha)/(t_purch+2*abs(tmean_u-t_)+beta)
    return lamb, tmean_u, k
```

```python
current = pd.to_datetime('2018-08-01')
```

```python
#test_id를 지정

test_id = bbfood_test.m_id.unique()[:10]
```

```python
func_lamb(train = bbfood_train, test = bbfood_test, test_id=test_id[0], alpha = alpha, beta=beta, tmean=tmean, current=current)
```

```python
current_l = pd.date_range(current, periods=30).tolist()
```

```python
current_l
```

```python
for i in range(10):
    print(func_lamb(train = bbfood_train, test = bbfood_test, test_id=test_id[0], 
              alpha = alpha, beta=beta, tmean=tmean, current=current_l[i]))
    
```
